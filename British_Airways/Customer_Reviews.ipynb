```python
# importing required libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
base_url = 'https://www.airlinequality.com/airline-reviews/british-airways/?sortby=post_date%3ADesc&pagesize=100'

# Requesting and Collecting the page data
page = requests.get(base_url)

# converting the page to html or parse content
parsed_content = BeautifulSoup(page.text, 'html')
```

```python
# Extracting first customer review
individual_customer = parsed_content.find('div', class_="body") 
```

```python
# Title of first customer review

individual_customer.find('h2').get_text()
```

```python
# Name country and date of first customer review

individual_customer.find('h3').get_text().strip()
```

```python
# Authentication of the trip of first customer

individual_customer.find('div',class_="text_content").get_text().strip().split('|')[0]
```

```python
# Function to extract title

def h2_tag(indi_customer):
    return indi_customer.find('h2').get_text()
```

```python
# Funtion to extract Name country and date

def h3_tag(indi_customer):
    return indi_customer.find('h3').get_text().strip()
```

```python
# Func to extract Authentication of the trip

def text_content(indi_customer):
    return indi_customer.find('div',class_="text_content").get_text().strip().split('|')[0]
```

```python
# Column headings of Data frame

col = ['Title','Name','Authentication']
```

```python
df = pd.DataFrame(columns=col)
df
```

```python
# Extracting all the customer reviews and parsing to htnml

all_customers = parsed_content.find_all('div',class_="body")
all_customers
```

```python
# looping over all the customer reviews and extracting the desired data

for each in all_customers:
    rows = []
    rows.append(h2_tag(each))
    rows.append(h3_tag(each))
    rows.append(text_content(each))
#     print(rows)
    
    length = len(df)
    df.loc[length] = rows
```

```python
# Before_Cleaning = df
Before_Cleaning
```

```python
# splitting the Name column to Name Counbtry and Date

df[['Name','Country','Date']]= df['Name'].str.extract(r'(.+?) \((.+?)\) (.+)')   # '\( \)' is to skip closed paranthesis
df
```

```python
# Removing unnecessary chars 

df['Title'] = df['Title'].str.replace('"','')
```

```python
df
```

```python
# Removing unnecessary data

df.Authentication = df['Authentication'].str.replace(r'[^a-xA-Z\s]+',"")
```

```python
# Converting Date column to 'date' format

df.Date = pd.to_datetime(df.Date)
```
